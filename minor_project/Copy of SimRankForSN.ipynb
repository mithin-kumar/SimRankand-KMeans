```python
def SimCluster(similarity,initial):
  cells ={}
  
  for center in initial:
#     idx=center
    cells[center]=[center]
  n = len(similarity)
  try :
      for node in range(n):
        max=0.0
        idx=list(initial)[0]
        for center in initial:
          if max < similarity[node][center]:
            idx= center
            max= similarity[node][center]
        cells[idx].append(node)
  except  KeyError:
    pass

  return cells
```

```python
from more_itertools import ilen

def SimRank(graph,similarity,d):
  score = {}
  for node in graph.nodes:
    sum =0
    # print(node)
    for v in graph.neighbors(node):
      sum += similarity[node][v]

    sum /= ilen(graph.neighbors(v))
    sum *= d
    sc = 1-d + sum

    score[node]=sc

  return score
```
